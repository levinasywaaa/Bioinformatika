```python
skor_siswa = float(input("Masukkan nilai siswa : "))
kehadiran = float(input("Masukkan persentase kehadiran siswa : "))
if skor_siswa >= 90 :
  if kehadiran >= 95 :
    print("Nilai A+ dengan Bonus")
  else :
    print("Nilai A")
elif skor_siswa >= 80 :
  if kehadiran >= 90 :
    print("Nilai B+")
  else :
    print("Nilai B")
elif skor_siswa >= 70 :
  print("Nilai C")
else :
  print("Tidak Lulus")
```
